```python
#array stack

class Stack_Arr:
    def __init__(self,n):
        self.top=-1
        self.stack=[0]*n
        
    def push(self,a):
        if self.top==len(self.stack)-1:
            print('No space in stack')
        else:
            self.top+=1
            self.stack[self.top]=a
            
    def pop(self):
        self.stack[self.top]=0
        self.top-=1
    
    def peek(self):
        return self.stack[self.top]
```

```python
#list stack

class Node:
    def __init__(self,elem,n):
        self.elem=elem
        self.next=n
        
class Stack_LL:
    def __init__(self):
        self.head=None
        self.top=-1
        
    def push(self,e):
        if self.head==None:
            self.head=Node(e,None)
        else:
            tail=Node(e,None)
            tail.next=self.head
            self.head=tail
        self.top+=1
        
    def pop(self):
        newhead=self.head.next
        self.head=None
        self.head=newhead
        self.top-=1 
        
    def peek(self):
        return self.head.elem
        
```

```python
#queue array
#last position=(st+sz-1)%len
#next available position=(st+sz)%len

class Queue_Arr:
    def __init__(self,n):
        self.queue=[0]*n
        self.start=0
        self.size=0
    
    def enqueue(self,a):
        if self.start==0:
            self.queue[self.start]=a
        else:
            next_av=(self.start+self.size)%len(self.queue)
            self.queue[self.next_av]=a
        self.size+=1
            
    def dequeue(self):
        self.queue[self.start]=0
        self.start=(self.start+1)%len(self.queue)
        self.size-=1
        
    def peek(self):
        return self.queue[self.start]
    
```

```python
#Queue Linked List
class Node:
    def __init__(self,elem,n,p):
        self.elem=elem
        self.next=n
        self.prev=p

class Queue_LL:
    def __init__(self):
        self.head=Node(None,None,None) #dummy head
        
    def enqueue(self,a):
        n=Node(a,None,None)
        if self.head.next==None:
            self.head.next=n
            n.prev=self.head
            self.head.prev=n
            n.next=self.head
        else:
            tail=self.head.prev
            tail.next=n
            n.prev=tail
            n.next=self.head
            self.head.prev=n
        
    def dequeue(self):
        start=self.head.next
        self.head.next=start.next
        start.next.prev=self.head
        
    def peek(self):
        return self.head.next.elem
        
```
